```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/housing.csv')
np_arr = df[['price', 'area']].to_numpy()
```

$$
a = \frac{\sum_{i=1}^{n}{(x_i - \bar{x})(y_i - \bar{y})})}{\sum_{i=1}^{n}({x_i - \bar{x}})^2}
$$

$$
b = \bar{y} - a\bar{x}
$$

```python
x = np_arr[:, 1]
y = np_arr[:, 0]
x[:5], y[:5]
```

```python
x_avg = np.average(x)
y_avg = np.average(y)
x_avg, y_avg
```

```python
a = np.sum((x - x_avg)*(y - y_avg)) / np.sum((x - x_avg)**2)
a
```

```python
b = y_avg - a * x_avg
b
```

```python
y_pred = a * x + b
y_pred
```

```python
plt.scatter(x, y, label='Actual Data')
plt.plot(x, y_pred, c="red", label="Regression Line", linewidth=3)
plt.xlabel("Area")
plt.ylabel("Price")
plt.legend()
plt.show()
```

$$
{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}|
$$

```python
mae = 1/len(x) * np.sum(np.abs(y - y_pred))
mae
```

$$
MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y})^2
$$

```python
mse = 1/len(x) * np.sum((y - y_pred)**2)
mse
```

$$
R^2 = 1 - \frac{MSE}{\frac{1}{n} \sum(y_i - \bar{y})^2}
$$

```python
r_2 = 1 - (mse / (1/len(x) * np.sum((y - y_avg)**2)))
r_2
```
