## Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt
```

$$
y = 3x^2 + 5x + 5
$$

```python
def f(x):
    return 3*x**2 + 5*x + 5

def f_grad(x):
    return 6*x + 5
```

```python
X = np.linspace(-10, 10, 100)
y = f(X)
y
```

```python
x0 = np.random.randint(-10, 10)
y0 = f(x0)
plt.plot(X, y)
plt.scatter(x0, y0, c='r')

for _ in range(1000):
    x_grad = f_grad(x0)
    x0 = x0 - 0.01 * x_grad
    
    y0 = f(x0)
    plt.scatter(x0, y0, c='g')

    if abs(x0) == abs(-0.84):
        print(f"Stop after {_}-step")
        break


plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
