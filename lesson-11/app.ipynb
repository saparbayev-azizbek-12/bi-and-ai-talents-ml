```python
from graphviz import Source
from sklearn.datasets import load_iris
from decision_tree.graph import draw_custom_tree
from decision_tree.classsifier import DecisionTreeClassifierCustom
from pprint import pprint

iris = load_iris(as_frame=True)
X = iris.data
y = iris.target

clf = DecisionTreeClassifierCustom(max_depth=4)
clf.fit(X.values, y.values)
pprint(clf.tree)

dot = draw_custom_tree(clf.tree, feature_names=iris.feature_names)
dot.render("source/custom_tree", format="dot", cleanup=False)

Source.from_file("source/custom_tree.dot")
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.tree import DecisionTreeRegressor
from decision_tree.regressor import DecisionTreeRegressorCustom

tree_clf = DecisionTreeRegressor(max_depth=4, random_state=42)
tree_clf.fit(X, y)
pred1 = tree_clf.predict(X)

tree_clf2 = DecisionTreeRegressorCustom(max_depth=4)
tree_clf2.fit(X, y)
pred2 = tree_clf2.predict(X)


mse_sklearn = mean_squared_error(y, pred1)
mse_custom = mean_squared_error(y, pred2)

print("Sklearn Tree MSE:", mse_sklearn)
print("Custom Tree MSE:", mse_custom)
```
