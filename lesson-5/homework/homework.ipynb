## Polynomial Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
m = 1000
X = 5 * np.random.rand(m,1) - 3
y = 0.5 * X**2 +  X + 10 + np.random.rand(m,1)
y
```

```python
plt.figure(figsize=[10,6])
plt.plot(X, y, 'b.')
plt.axis([-3, 2, 9, 13])
plt.grid()
plt.show()
```

```python
X2 = X ** 2

X_all = np.c_[X, X2]

X_all
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_all, y)
y_pred = lin_reg.predict(X_all)
```

```python
plt.plot(X,y, 'b.')
plt.scatter(X, y_pred, c='red')
plt.grid()
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_poly, y)
```

```python
X_test = np.linspace(X.min(), X.max(), 500).reshape(-1, 1)
X_test_poly = poly.transform(X_test)
y_pred = model.predict(X_test_poly)
```

```python
plt.plot(X, y, 'b.')
plt.plot(X_test, y_pred, c='red', linewidth='3')
plt.show()
```
